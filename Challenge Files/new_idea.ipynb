```python
ccms_fault =[]
slid_fault = []
for index,row in df.iterrows():
    if row.power >= (df.power.mean()+df.power.std()) or row.power <= (df.power.mean()-df.power.std()):
        ccms_fault.append(row.ccms)
        slid_fault.append(row.slid)
```
